```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://talentcocomedia.s3.amazonaws.com/ml-assets/titanic.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
data = df.copy()
```

```python
data=df.fillna({'Age': df['Age'].mean()})
```

```python
data.isnull().sum(axis=0)
```

```python
data = data.drop(['Cabin'], axis =1)
```

```python
data.head()
```

```python
data = data.dropna()
```

```python
data.info()
```

```python
data = data.drop(['PassengerId','Name','Ticket','Embarked'], axis=1)
```

```python
data.head()
```

```python
data['Sex'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
sex_enc=LabelEncoder()
```

```python
data['Sex'] = sex_enc.fit_transform(data['Sex'])
```

```python
data.head()
```

```python
data['Sex'].unique()
```

```python
data =  pd.get_dummies(data, columns = ['Sex'])
```

```python
data.head()
```

```python
sns.heatmap(data.corr(), annot = True)
```

```python
sns.pairplot(data)
```

```python
plt.scatter(data[['Age']], data[['Survived']])
```

```python
sns.pairplot(data, hue='Survived')
```

```python
data.drop(['SibSp','Parch'], axis=1)
data.head()
```

```python
data['Survived'].value_counts()
```

```python
x = data.drop(['Survived'],axis=1)
x.head()
```

```python
y = data[['Survived']]
y.head()
```

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression(penalty='l2', solver= 'newton-cg', C=5)
```

```python
lr_model.fit(x_train,y_train)
```

```python
y_pred_lr = lr_model.predict(x_test)
y_pred_lr
```

```python
from sklearn.metrics import accuracy_score
acc_lr = accuracy_score(y_pred_lr,y_test)
acc_lr
```
